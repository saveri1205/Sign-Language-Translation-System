```python
#select a random image
def path_of_image(current_letter):
  path = r"/content/drive/MyDrive/DATA/train/"+current_letter
  random_filename = random.choice([
      x for x in os.listdir(path)
      if os.path.isfile(os.path.join(path, x))
  ])
  path=path+"/"+random_filename
  print(path)
  return path
```

```python
@anvil.server.callable
def anvil_translate(text):
  input_string=text
  list_of_characters=list(input_string)
  img_array = []
  print("individual characters = ",list_of_characters)

  for i in list_of_characters:

    if(i==" "):
      continue

    cur_image_path=path_of_image(i)
    img = cv2.imread(cur_image_path)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)


  out = cv2.VideoWriter('/content/drive/MyDrive/capstone_demo/project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 15, size)
  
  for i in range(len(img_array)):
      out.write(img_array[i])
  out.release()

  clip = VideoFileClip("/content/drive/MyDrive/capstone_demo/project.avi")

    
  # applying speed effect
  final = clip.fx( vfx.speedx, 0.1)
  print(type(final))
    
  # showing final clip
  final.ipython_display()


  final.write_videofile("/content/drive/MyDrive/capstone_demo/project.avi",codec="libx264")

  return "https://drive.google.com/drive/folders/1J9skRYU-DFUk6t2-bLjyJvkg3ZwE_2fK"
```
